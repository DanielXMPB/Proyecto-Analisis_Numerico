```python
import matplotlib.pyplot as plt 
import scipy.io as sio
from spec2rgb import ColourSystem

#from scipy.ndimage.filters import gaussian_filter
import numpy as np
#import cv2

data_path = '/content/drive/My Drive/Anlisis numrico/JasperRidge.mat'
var_name = 'Z'
img_spec = sio.loadmat(data_path)[var_name]

color_space = "sRGB"
start, end = 380, 2500 # VariSpec VIS
number_bands = 198

cs = ColourSystem(cs=color_space, start=start, end=end, num=number_bands)
img_rgb = cs.spec_to_rgb(img_spec)
img_rgb = img_rgb/np.max(img_rgb)

plt.imshow(img_rgb)
plt.show()

#blurred = gaussian_filter(img_rgb, sigma=(0.75,0,0.75))
#plt.imshow(blurred)
#plt.show()

#res = np.reshape(img_rgb,(25,25,198))
#plt.imshow(res)
#plt.show()

#res = cv2.resize(img_rgb, dsize=(25,25), interpolation=cv2.INTER_CUBIC)
#plt.imshow(res)
#plt.show()
```
